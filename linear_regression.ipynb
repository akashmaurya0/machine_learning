```python
import numpy as np
import matplotlib.pyplot as plt

# Input data (hours studied)
X = np.random.rand(100,1)

# Output data (marks obtained)
Y = 4+3*X+np.random.rand(100,1)

# Initialize parameters
b0 = 0  # Intercept
b1 = 0  # Slope
learning_rate = 0.01
epochs = 1000  # Number of iterations

# Step 4: Cost function (MSE)
def mean_squared_error(Y_true, Y_pred):
    return np.mean((Y_true - Y_pred) ** 2)

# Step 5: Gradient descent function
def gradient_descent1(X, Y, b0, b1, learning_rate, epochs):
    n = len(X)  # Number of data points
    cost_history = []  # To store the cost at each epoch
    
    for i in range(epochs):
        # Step 1: Make predictions using the current parameters
        Y_pred = b0 + b1 * X
        
        # Step 2: Calculate the cost (MSE)
        cost = mean_squared_error(Y, Y_pred)
        cost_history.append(cost)
        
        # Step 3: Compute the gradients (partial derivatives)
        db0 = -(2/n) * sum(Y - Y_pred)  # Gradient of the intercept
        db1 = -(2/n) * sum((Y - Y_pred) * X)  # Gradient of the slope
        
        # Step 4: Update the parameters using the gradients
        b0 = b0 - learning_rate * db0
        b1 = b1 - learning_rate * db1
        
        # Optionally, print the cost every 100 epochs
        if i % 100 == 0:
            print(f'Epoch {i}: Cost = {cost}')
    
    return b0, b1, cost_history

# Train the model using gradient descent
b0, b1, cost_history = gradient_descent1(X, Y, b0, b1, learning_rate, epochs)

# Make predictions using the final slope and intercept
Y_pred = b0 + b1 * X

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual data')

# Plot the regression line
plt.plot(X, Y_pred, color='red', label='Regression line')

plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.legend()
plt.show()

# Plot the cost function history
plt.plot(range(epochs), cost_history)
plt.xlabel('Epoch')
plt.ylabel('Cost (MSE)')
plt.title('Cost Function History')
plt.show()
```
